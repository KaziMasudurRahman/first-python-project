```
data = []
choice = 'random'

def line():
    print('---------------------------')

def show_course():
    line()
    print('Courses list')
    line()
    print('1. Enroll Course')
    print('2. unenroll Course')
    print('3. View Current Courses list')
    print('4. Exit')
    line()
    choice = input('Enter your choice number: ')
    return choice

while choice != '4':
    choice = show_course()
    if choice == '1':
        item = input('Course Name: ')
        data.append(item)
        print('"',item,'"',' is added in the list')
    elif choice == '2':
        item = input('Unenroll course name: ')
        if item in data:
            data.remove(item)
            print('Removed ', '"',item,'"')
        else:
            print('Could not find ','"',item,'"',' course')
    elif choice == '3':
        print('List of Current Courses:')
        for item in data:
            print(item)
    elif choice == '4':
        print('Goodbye!')
    else:
        print('Please enter one of 1, 2, 3 or 4')
```
